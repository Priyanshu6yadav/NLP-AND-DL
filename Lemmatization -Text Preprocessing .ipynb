## ðŸŒ¿ WordNet Lemmatization (Easy Explanation)

**WordNet lemmatization** reduces a word to its correct *dictionary form* (lemma) using the WordNet lexical database.  
Unlike stemming, it always produces **real English words** by considering grammar rules and word meaning. Which is a root words rather then root stem, the output of stemming. after lemmatization, we will be getting a valid word that means the same thing.

### Example:

| Word     | Lemma |
|----------|--------|
| running  | run    |
| studies  | study  |
| better   | good   |
| cars     | car    |

Lemmatization = **clean, meaning-based word reduction**.

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
'''
POS - Noun-n
verb-v
adjective-a
adverb-r
'''
lemmatizer.lemmatize("going",pos='r')
```

```python
words = ['eating','eats','eaten','writing','writes','programming','programs','history','finally','finalized']
```

```python
for word in words:
    print(word +" "
    "---> "+lemmatizer.lemmatize(word,pos='v'))
## pos : str
## The Part Of Speech tag. Valid options are "n" for nouns, "v" for verbs, "a" for adjectives, "r" for adverbs and "s" for satellite adjectives.
```

```python
lemmatizer.lemmatize('fairly',pos='v'),lemmatizer.lemmatize("sporting")
```
