```python
## Create average Embedding
import numpy as np

def avg_embedding(tokens):
    # This function calculates the average Word2Vec embedding for a list of tokens (words) in a review.
    # It iterates through each token, retrieves its vector from the trained Word2Vec model (`model.wv`).
    # It filters out words that are not present in the Word2Vec model's vocabulary (out-of-vocabulary words).
    valid = [model.wv[word] for word in tokens if word in model.wv]
    if len(valid) == 0:
        # If no valid words are found (e.g., all words are out-of-vocabulary), return a zero vector of the specified embedding size.
        return np.zeros(model.vector_size)
    # Otherwise, compute the mean of all valid word vectors to get a single average embedding for the review.
    return np.mean(valid, axis=0)

# Apply the avg_embedding function to each review in X_train and X_test to convert text reviews into numerical vector representations.
# X_train_w2v and X_test_w2v will now contain the averaged Word2Vec embeddings for the training and testing datasets, respectively.
X_train_w2v = np.array([avg_embedding(tokens.split()) for tokens in X_train])
X_test_w2v = np.array([avg_embedding(tokens.split()) for tokens in X_test])
```

```python
import pandas as pd

def parse_classification_report(report_string):
    lines = report_string.split('\n')
    metrics = {}
    for line in lines:
        line = line.strip()
        if not line:
            continue

        parts = line.split()
        if not parts:
            continue

        # Check for class-specific metrics (lines starting with a digit)
        if parts[0].isdigit() and len(parts) >= 5:
            label = parts[0]
            metrics[f'{label}_precision'] = float(parts[1])
            metrics[f'{label}_recall'] = float(parts[2])
            metrics[f'{label}_f1-score'] = float(parts[3])
            metrics[f'{label}_support'] = int(parts[4])
        # Check for accuracy, macro avg, weighted avg
        elif 'accuracy' in line and 'accuracy' not in metrics: # Ensure not to overwrite if explicitly set later
            try:
                metrics['accuracy'] = float(parts[-2]) # Accuracy is typically second to last
            except ValueError:
                pass # If parsing fails, ignore and let explicit setting handle it
        elif 'macro avg' in line and len(parts) >= 5:
            metrics['macro_avg_precision'] = float(parts[2])
            metrics['macro_avg_recall'] = float(parts[3])
            metrics['macro_avg_f1-score'] = float(parts[4])
        elif 'weighted avg' in line and len(parts) >= 5:
            metrics['weighted_avg_precision'] = float(parts[2])
            metrics['weighted_avg_recall'] = float(parts[3])
            metrics['weighted_avg_f1-score'] = float(parts[4])
    return metrics

# SVM Metrics (from cell 1-TjaeQEQ9TM)
svm_report_string = """              precision    recall  f1-score   support

           0       0.00      0.00      0.00       793
           1       0.67      1.00      0.80      1607

    accuracy                           0.67      2400
   macro avg       0.33      0.50      0.40      2400
weighted avg       0.45      0.67      0.54      2400"""
svm_metrics = parse_classification_report(svm_report_string)
svm_metrics['accuracy'] = 0.6695833333333333 # Direct accuracy from output

# Logistic Regression Metrics (from cell df1364a0)
log_reg_report_string = """               precision    recall  f1-score   support

           0       0.67      0.47      0.55       775
           1       0.78      0.89      0.83      1625

    accuracy                           0.75      2400
   macro avg       0.72      0.68      0.69      2400
weighted avg       0.74      0.75      0.74      2400"""
log_reg_metrics = parse_classification_report(log_reg_report_string)
log_reg_metrics['accuracy'] = 0.7545833333333334 # Direct accuracy from output

# Random Forest Metrics (from cell eae40650)
rf_report_string = """Random Forest Accuracy: 0.7529166666666667

Random Forest Classification Report:
               precision    recall  f1-score   support

           0       0.65      0.50      0.57       775
           1       0.79      0.87      0.83      1625

    accuracy                           0.75      2400
   macro avg       0.72      0.69      0.70      2400
weighted avg       0.74      0.75      0.74      2400"""
# Clean the string for Random Forest as it includes extra line before the classification report.
rf_report_string_clean = "\n".join(rf_report_string.split('\n')[3:])
rf_metrics = parse_classification_report(rf_report_string_clean)
rf_metrics['accuracy'] = 0.7529166666666667 # Direct accuracy from output

# Create a DataFrame for comparison
comparison_df = pd.DataFrame.from_dict({
    'SVM': svm_metrics,
    'Logistic Regression': log_reg_metrics,
    'Random Forest': rf_metrics
}, orient='index')

print("Model Performance Comparison:\n")
print(comparison_df[['accuracy', '0_precision', '0_recall', '0_f1-score', '1_precision', '1_recall', '1_f1-score', 'macro_avg_f1-score', 'weighted_avg_f1-score']].round(2))

# Textual Summary
print("\n--- Model Performance Summary ---")
print("Overall, Logistic Regression and Random Forest models performed significantly better than SVM.")
print(f"Logistic Regression achieved the highest accuracy at {comparison_df.loc['Logistic Regression', 'accuracy']:.2f}.")
print(f"Random Forest was very close with an accuracy of {comparison_df.loc['Random Forest', 'accuracy']:.2f}.")
print("The SVM model performed poorly, especially for class 0 (negative reviews), where its precision, recall, and F1-score were 0.00. This indicates it failed to correctly predict any negative reviews.")
print("Logistic Regression and Random Forest show balanced performance across both classes, with Logistic Regression having slightly better precision for class 0.")
print("For future steps, fine-tuning Logistic Regression or Random Forest, or exploring other advanced models, would be beneficial.")
```
