# ðŸŒ± What is Stemming in NLP?

**Stemming** is the process of reducing a word to its **root form** (called â€œstemâ€).  
The stem may **not** be a real English word â€” itâ€™s just a simplified form.

### Example:

| Word    | Stem                                      |
|---------|-------------------------------------------|
| playing | play                                      |
| played  | play                                      |
| player  | player (or playr depending on stemmer)    |
| studies | studi                                     |

Stemming helps because many forms of a word have the same meaning in text.

# ðŸ”¥ Why use Stemming?

To make text simpler for models by reducing variations of words.

### Example:

â€œplayâ€, â€œplayingâ€, â€œplayedâ€ â†’ **â€œplayâ€**

This reduces noise and improves search, analysis, and text mining.

### Types of Stemming
There are 3 popular types:

#### 1ï¸âƒ£ Porter Stemmer (most common)
	â€¢	Classical and widely used
	â€¢	Good balance between speed and accuracy
	â€¢	Produces stems that may not always be meaningful words


```python
words = ['eating','eats','eaten','writing','writes','programming','programs','history','finally','finalized']
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word+" ----> "+stemming.stem(word))
```

```python
## major major Disadvantage of stemming --history---->histori
stemming.stem("Congratulations")
```

```python
stemming.stem("sitting")
```

### RegexpStemmer class
* RegexpStemmer is an NLTK stemmer that removes word endings using regular expressions.
* You define a regex pattern (like "ing$"), and it strips that ending from words.
* Example: "playing" â†’ "play" using pattern "ing$".

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min =4) ## if in last word this value is present --> ing$|s$|e$|able$ remove it
```

```python
reg_stemmer.stem('eating') # remove --> ing
```

```python
reg_stemmer.stem('playing') # remove --> ing
```

```python
reg_stemmer.stem('sleeping') # remove --> ing
```

```python
reg_stemmer.stem('applycable') ## Remove --> able
```

### 3ï¸âƒ£ Lancaster Stemmer (Aggressive)
	â€¢	Much more strict and aggressive
	â€¢	Often cuts too much
	â€¢	Stems can look unnatural

```python
from nltk.stem import LancasterStemmer
ls = LancasterStemmer()
words = ["problem", "problems", "problematic"]
for w in words:
    print(w, "->", ls.stem(w))
```
