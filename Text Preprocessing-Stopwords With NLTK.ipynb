```python
## Speech of DR APJ Abdul Kalam
# Define a multi-line string variable 'paragraph' containing the speech content.
paragraph = """I have three visions for India. In 3000 years of our history, people from all over
               the world have come and invaded us, captured our lands, conquered our minds.
               From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British,
               the French, the Dutch, all of them came and looted us, took over what was ours.
               Yet we have not done this to any other nation. We have not conquered anyone.
               We have not grabbed their land, their culture,
               their history and tried to enforce our way of life on them.
               Why? Because we respect the freedom of others.That is why my
               first vision is that of freedom. I believe that India got its first vision of
               this in 1857, when we started the War of Independence. It is this freedom that
               we must protect and nurture and build on. If we are not free, no one will respect us.
               My second vision for India’s development. For fifty years we have been a developing nation.
               It is time we see ourselves as a developed nation. We are among the top 5 nations of the world
               in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling.
               Our achievements are being globally recognised today. Yet we lack the self-confidence to
               see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect?
               I have a third vision. India must stand up to the world. Because I believe that unless India
               stands up to the world, no one will respect us. Only strength respects strength. We must be
               strong not only as a military power but also as an economic power. Both must go hand-in-hand.
               My good fortune was to have worked with three great minds. Dr. Vikram Sarabhai of the Dept. of
               space, Professor Satish Dhawan, who succeeded him and Dr. Brahm Prakash, father of nuclear material.
               I was lucky to have worked with all three of them closely and consider this the great opportunity of my life.
               I see four milestones in my career"""
```

```python
# Import the stopwords module from the NLTK corpus.
from nltk.corpus import stopwords
```

```python
# Retrieve and display the list of English stopwords.
stopwords.words('english') # this all are --> stopwords
```

```python
# Retrieve and display the list of Russian stopwords.
stopwords.words('russian')
```

```python
# Import the PorterStemmer from the NLTK stem module.
from nltk.stem import PorterStemmer
import nltk
```

```python
# Initialize an instance of the PorterStemmer.
stemmer = PorterStemmer()
```

```python
# Tokenize the 'paragraph' into a list of sentences. This process is called tokenization.
sentences=nltk.sent_tokenize(paragraph) ## paragraph --> sentences this process are called toknization
```

```python
# Check the data type of the 'sentences' variable.
type(sentences)
```

```python
## Apply Stopwords And Filter And then Apply Stemming

# Iterate through each sentence to apply tokenization, stopword removal, and Porter Stemming.
for i in range(len(sentences)):
    # Tokenize the current sentence into words.
    words=nltk.word_tokenize(sentences[i])
    # Stem each word and filter out English stopwords.
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    # Join the processed words back into a sentence.
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```

```python
# Display the sentences after applying Porter Stemming and stopword removal.
sentences ## All stopword is remove
```

```python
# Import the SnowballStemmer from the NLTK stem module.
from nltk.stem import SnowballStemmer
```

```python
# Initialize an instance of the SnowballStemmer for English language.
snowballstemmer = SnowballStemmer('english')
```

```python
## Apply Stopwords And Filter And then Apply Snowball Stemming
# Iterate through each sentence to apply tokenization, stopword removal, and Snowball Stemming.
# Note: This is applied to the already Porter-stemmed 'sentences' from the previous step.
for i in range(len(sentences)):
    # Tokenize the current sentence into words.
    words=nltk.word_tokenize(sentences[i])
    # Stem each word using Snowball Stemmer and filter out English stopwords.
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    # Join the processed words back into a sentence.
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```

```python
# Display the sentences after applying Snowball Stemming. Note that capitalization is removed during this process.
sentences ## After applying snowball stemmer all the captial letter become smaller
```

```python
# Import the WordNetLemmatizer from the NLTK stem module.
from nltk.stem import WordNetLemmatizer
```

```python
# Initialize an instance of the WordNetLemmatizer.
lemmatizer = WordNetLemmatizer()
```

```python
# Apply WordNet Lemmatization to the sentences.
# Note: This code attempts to lemmatize words that have already been stemmed by Porter and Snowball stemmers.
# This can lead to less meaningful results as stemmers often reduce words to non-dictionary forms.
for i in range(len(sentences)):
    # Tokenize the current sentence into words.
    words=nltk.word_tokenize(sentences[i])
    # Lemmatize each word (converted to lowercase, treated as a verb) and filter out stopwords.
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    # Join the processed words back into a sentence.
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```

```python
# Display the sentences after attempting WordNet Lemmatization on the already stemmed text.
# As noted previously, the output might not be fully meaningful due to prior stemming.
sentences
```

### Correcting Lemmatization: Applying to original text

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# Re-tokenize sentences from the original paragraph to get fresh input for lemmatization.
# This ensures lemmatization is applied to un-stemmed text for more accurate results.
sentences_for_lemmatization = nltk.sent_tokenize(paragraph)

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

lemmatized_sentences = []
for sentence in sentences_for_lemmatization:
    words = nltk.word_tokenize(sentence)
    # Lemmatize words, convert to lowercase, and filter stopwords.
    # Using pos='v' for verbs, you can also try 'n' for nouns or leave it default if part-of-speech tagging is not available.
    # 'word.isalpha()' is added to filter out punctuation and numbers before lemmatization.
    lemmatized_words = [lemmatizer.lemmatize(word.lower(), pos='v') for word in words if word.lower() not in stop_words and word.isalpha()]
    lemmatized_sentences.append(' '.join(lemmatized_words))

# Display the lemmatized sentences.
print("Lemmatized sentences (from original paragraph, removing non-alphabetic tokens and stopwords):")
for s in lemmatized_sentences:
    print(s)
```

In the code above:
- I recreated `sentences_for_lemmatization` from the original `paragraph` to ensure we are working with un-stemmed text.
- I also added `word.isalpha()` to filter out punctuation and numbers, and used `word.lower()` before checking against stopwords and lemmatizing to ensure consistency.
- The `pos='v'` argument tells the lemmatizer to treat words as verbs, which often gives better results for many words. You might experiment with `pos='n'` for nouns or remove the `pos` argument to let it default to nouns if verb lemmatization isn't appropriate for all words.
