1. Corpus    --> paragraph
2. Documents --> sentence
3. Vocabulary --> unique word in paragraph
4. Words      --> all words which are present in corpus

```python
corpus = """ Hello welcome, to 'priyanshu githud nlp and dl code.
Please do see the entire code! to become expert in NLP. """
```

```python
corpus
```

#### Tokenization 

```python
##. paragraph ---> sentence
from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
## Toknization
## Paragraph --> words
## Sentence  --> words
from nltk import word_tokenize
```

```python
word_tokenize(corpus)  ## paragraph --> words ## My paragraph is toknized into words
## Why we do this because each and every words have there difference
## And we perform some preprocessing, clean it and somany things
```

```python
for sentence in documents:
    print(word_tokenize(sentence))   ## sentence --> word 
```

```python
from nltk.tokenize import wordpunct_tokenize
## Punct is also treated as word
```

```python
wordpunct_tokenize(corpus)  # older -> 'to',
# "'priyanshu",
# 'githud',
# after apply wordpunct_tokenize --> 'to',
# "'",
# 'priyanshu',
# 'githud',
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
TreebankWordTokenizer().tokenize(corpus)
### * split standard contractions, e.g. don't -> do n't and they'll -> they 'll
#.  * treat most punctuation characters as separate tokens
#.  * split off commas and single quotes, when followed by whitespace
#.  * separate periods that appear at the end of line
```
