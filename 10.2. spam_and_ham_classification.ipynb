```python
import pandas as pd

messages = pd.read_csv(
    "/content/SMSSpamCollection.txt",
    sep="\t",
    names=["label", "message"],
)
```

```python
messages
```

```python
## Data cleaning and preprocessing
import re
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmitizer = WordNetLemmatizer()
```

```python
## Applying lemmatizer
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmitizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

### Train Test Split Dataset

```python
y = pd.get_dummies(messages['label'])
y = y.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y, test_size=0.20)
```

### Trained our model by train test data


#### Create Bag Of Word

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X_train = cv.fit_transform(X_train)
X_test = cv.transform(X_test)
```

```python
import numpy as np

# Display large arrays without truncation
np.set_printoptions(
    edgeitems=30,        # Show 30 items at the start and end
    linewidth=100000,    # Very wide line so arrays don't wrap
    formatter={'float': lambda x: "%.3g" % x}   # Format float values
)
X_train
```

```python
cv.vocabulary_
```

```python
## Trained model
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

## Create TF_IDE

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(corpus,y, test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train = tv.fit_transform(X_train).toarray()
X_test = tv.transform(X_test).toarray()
```

```python
X_train
```

```python
tv.vocabulary_
```

```python
## Trained model
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train,y_train)
y_pred = spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
